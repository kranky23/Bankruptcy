```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

in_file='bank.csv'
text = pd.read_csv(in_file)
text.columns = ["Industrial_Risk","Mgmt_Risk","Financial_Flexibility","Credibility","Competitiveness","Operating_Risk","Classifier"]
text.head()
```

```python
text['Industrial_Risk'] = text['Industrial_Risk'].map({'N': -1,'A': 0,'P': 1})
text['Mgmt_Risk'] = text['Mgmt_Risk'].map({'N': -1,'A': 0,'P': 1})
text['Financial_Flexibility'] = text['Financial_Flexibility'].map({'N': -1,'A': 0,'P': 1})
text['Credibility'] = text['Credibility'].map({'N': -1,'A': 0,'P': 1})
text['Competitiveness'] = text['Competitiveness'].map({'N': -1,'A': 0,'P': 1})
text['Operating_Risk'] = text['Operating_Risk'].map({'N': -1,'A': 0,'P': 1})
text['Classifier'] = text['Classifier'].map({'B': 1,'NB' :0})
text.head()
# data['parents'] = data['parents'].map({'usual': 0 ,'pretentious': 1 ,'great_pret': 2})
```

```python
# text.hist()
# text.boxplot()
# plt.pie(text['Financial Flexibility'],text['Operating Risk'])
text.Operating_Risk.value_counts().plot(kind='pie',autopct = '%1.1f%%',colors=['r','m','g'])
plt.axis('equal') 
plt.title('Operating_Risk') 
fig = plt.gcf() 
fig.set_size_inches(5,5) 
plt.legend()
plt.show() 
```
