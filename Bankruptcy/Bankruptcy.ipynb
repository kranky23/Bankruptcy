# Reading the data and naming the columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

in_file='bank.csv'
text = pd.read_csv(in_file)
text.columns = ["Industrial_Risk","Mgmt_Risk","Financial_Flexibility","Credibility","Competitiveness","Operating_Risk","Classifier"]
text.head()
```

# Changing the characters to numbers

```python
#Converting the values of P, N, A to 1, -1, 0 respectively

text['Industrial_Risk'] = text['Industrial_Risk'].map({'N': -1,'A': 0,'P': 1})
text['Mgmt_Risk'] = text['Mgmt_Risk'].map({'N': -1,'A': 0,'P': 1})
text['Financial_Flexibility'] = text['Financial_Flexibility'].map({'N': -1,'A': 0,'P': 1})
text['Credibility'] = text['Credibility'].map({'N': -1,'A': 0,'P': 1})
text['Competitiveness'] = text['Competitiveness'].map({'N': -1,'A': 0,'P': 1})
text['Operating_Risk'] = text['Operating_Risk'].map({'N': -1,'A': 0,'P': 1})
# text['Classifier'] = text['Classifier'].map({'B': 1,'NB' :0})
text.head()
```

# Plotting Graphs


Now that the values have been changed from characters to integers, we can now plot different charts for the variables.
Let us start with pie charts.

```python
# text.hist()
# text.boxplot()
# plt.pie(text['Financial Flexibility'],text['Operating Risk'])

plt.figure(0)
t = text.Industrial_Risk.value_counts().plot(kind='pie',autopct = '%1.1f%%',colors=['r','g','b'])
plt.title('Industrial_Risk') 
fig = plt.gcf() 
fig.set_size_inches(4,4)
plt.legend()

plt.figure(1)
t = text.Mgmt_Risk.value_counts().plot(kind='pie',autopct = '%1.1f%%',colors=['r','g','b'])
plt.title('Mgmt_Risk') 
fig = plt.gcf() 
fig.set_size_inches(4,4)
plt.legend()

plt.figure(2)
text.Financial_Flexibility.value_counts().plot(kind='pie',autopct = '%1.1f%%',colors=['r','g','b'])
# plt.axis('equal') 
plt.title('Financial_Flexibility') 
fig = plt.gcf() 
fig.set_size_inches(4,4) 
plt.legend()


plt.figure(3)
text.Credibility.value_counts().plot(kind='pie',autopct = '%1.1f%%',colors=['r','g','b'])
# plt.axis('equal') 
plt.title('Credibility') 
fig = plt.gcf() 
fig.set_size_inches(4,4) 
plt.legend()


plt.figure(4)
text.Competitiveness.value_counts().plot(kind='pie',autopct = '%1.1f%%',colors=['r','g','b'])
# plt.axis('equal') 
plt.title('Competitiveness') 
fig = plt.gcf() 
fig.set_size_inches(4,4) 
plt.legend()


plt.figure(5)
text.Operating_Risk.value_counts().plot(kind='pie',autopct = '%1.1f%%',colors=['r','g','b'])
# plt.axis('equal') 
plt.title('Operating_Risk') 
fig = plt.gcf() 
fig.set_size_inches(4,4) 
plt.legend()


plt.show()

```

# Bar Graph

A sample of the bar graph representation is shown.

```python
plt.figure()
t = text.Classifier.value_counts().plot(kind='bar',colors=['r','g'])
plt.title('Classfier')
fig = plt.gcf()
fig.set_size_inches(4,4)
plt.show()
```

From the above data, we can infer that most of the candidates fall into the category of Non Bankruptcy

# Histograms

Variables under different categories are plotted using a histogram.

From the above histograms, it is shown that the variables not uniformly distributed. The number of values in the respective fields i.e. the positive, negative and average fields can be calculated and compared.
